In this task I make a Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly

```python
#import required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
import graphviz
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
import pydot
from IPython.display import Image
```

```python
# import data from source
iris = datasets.load_iris()

#assigning X(feature) and y(target)
X = iris.data
y = iris.target
```

```python
# decision tree model training
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1) 
tree = DecisionTreeClassifier()
tree.fit(X_train,y_train)
print("Training Complete.")
```

```python
# decision tree model prediciton vs. actual values
y_pred = tree.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}) 
df
```

```python
# visualize the graph
graphviz.Source(export_graphviz(tree,
                               out_file='img\desision_tree.dot',
                               feature_names = iris.feature_names,
                               class_names = iris.target_names,
                               filled=True,
                               rounded=True,
                               special_characters=True))

(graph,) = pydot.graph_from_dot_file('img\desision_tree.dot')
graph.write_png('img\desision_tree.png')

Image(filename='img\desision_tree.png')
```

Now, let us check our Decision tree assuming random data for a flower with values,
* sepal length (cm) = 4.2,
* sepal width (cm) = 2,
* petal length (cm) = 1.1,
* petal width (cm) = 0.1

```python
ip = np.array([4.2, 2, 1.1, 0.1]).reshape(1,-1)
output = tree.predict(ip)
print('The given data belongs to {}.'.format(iris.target_names[output[0]]))
```

Hence, our Decision Tree classifier can now classify new values too.
